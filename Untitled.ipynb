```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## LIT Exercise

This notebook walks through the given instructions. 


Begin by loading in the given data. Dates are harmonized across the two series, dropping market holidays. 

```python
ust = pd.read_csv('3MO.csv', index_col=0)
ust['DTB3'] = ust['DTB3'].replace('.', np.nan)
ust = ust.dropna(axis=0)
ust.index = pd.DatetimeIndex(ust.index)
ust['DTB3'] = pd.to_numeric(ust['DTB3'])


oil = pd.read_csv('RWTCd.csv', skiprows=2, index_col=0)
oil.index = pd.DatetimeIndex(oil.index)

idx = sorted(set(oil.index).intersection(set(ust.index)))
ust = ust.reindex(index=idx)
oil = oil.reindex(index=idx)
```

Part (A) asks to compare daily returns. The Treasury bill data are presented in annual percent return units, and the oil data give daily prices. To compare daily rates of return directly, some minor calculations are needed.


For the Treasury security, convert the annual return rate to a daily return rate, and recover a returns index. The interpretation of the index is that the percent change in the index over time reflects the returns to holding 3 month US Treasurys. The index is constructed such that December 31, 2019 = 100. 


For oil, construct a comparable price index by normalizing the daily price series to the price as of December 31, 2019. The interpretation of this index is that the percent change in the index over time reflects the change in oil prices, or the returns to an exposure to spot oil prices. 


Daily percent changes in the price of oil also follow directly from the price series. One special case for the oil price series arose on April 20, 2020, when pandemic-related circumstances led to negative spot prices. To retain an intuitive interpretation, the below calculation for daily return rates uses the absolute value of the lagged price in the denominator. In this way, a positive "percent change" indicates oil price appreciation, regardless of whether yesterday's spot rate was above zero. 

```python

trading_days = ust.resample('A').count().loc['2020-12-31', 'DTB3']  # usually ~252
# daily return rate
ust['DTB3_daily'] = (ust['DTB3'] / 100 + 1.) ** (1/trading_days) 
# price index
ust['DTB3_index'] = ust['DTB3_daily'].cumprod()
ust['DTB3_index'] = 100 * ust['DTB3_index'] / ust.loc['2019-12-31', 'DTB3_index']
# daily % change in price index
ust['DTB3_daily_pct'] = (ust['DTB3_daily'] - 1.) * 100


# daily return rate
oil['OIL_daily'] = (oil.iloc[:,0] - oil.iloc[:,0].shift(1)) / np.abs(oil.iloc[:,0].shift(1)) + 1
oil['OIL_daily'].iloc[0] = 1.
# daily pct change
oil['OIL_daily_pct'] = (oil['OIL_daily'] - 1.) * 100
# price index
oil['OIL_index'] = oil.iloc[:, 0] / oil.iloc[:, 0].loc['2019-12-31'] * 100


def lineChart(series, title, wide=False):
    if wide:
        fig, ax = plt.subplots(figsize=(10, 5))
    else:
        fig, ax = plt.subplots()
    ax.plot(series)
    ax.set_title(title)
    plt.show()
    plt.close()

def twoLineChart(series1, label1, series2, label2, title):
    fig, ax = plt.subplots()
    ax.plot(series1)
    ax.plot(series2)
    ax.set_title(title)
    ax.legend([label1, label2])
    plt.show()
    plt.close()

lineChart(ust.loc['2019-12-31':'2020-12-31','DTB3_index'], 
          'Index for 3M UST\n(2019-12-31=100)')
twoLineChart(ust.loc['2019-12-31':'2020-12-31','DTB3_index'], '3M UST',
             oil.loc['2019-12-31':'2020-12-31','OIL_index'], 'Oil',
             'Indexes for Oil & 3M UST\n(2019-12-31=100)')
    
```

## (A) Daily Excess Returns

The first chart below shows daily returns for both spot oil prices and the 3M UST. The units are percent change day-to-day. 


The second chart below answers part (A) of the instructions. Excess returns are defined as the oil percent return less the UST percent return. This reflects a spread between the daily returns to spot oil prices, versus holding 3M UST. Another possible interpretation is the balance of a long position in spot oil and a short position in 3M UST.

Please note the adjusted definition of "percent change" discussed above. 

```python
twoLineChart(ust.loc['2019-12-31':'2020-12-31','DTB3_daily_pct'], '3M UST',
             oil.loc['2019-12-31':'2020-12-31','OIL_daily_pct'], 'Oil', 
             'Daily Percent Change in Index\nOil & 3M UST')

## (A) daily excess returns
series = oil.loc['2019-12-31':'2020-12-31','OIL_daily_pct'] - ust.loc['2019-12-31':'2020-12-31','DTB3_daily_pct']
lineChart(series, 'Difference in Daily Percent Change in Index\nOil less 3M UST')
```

The charts above are dominated by the exceptional case of April 2020. To get a better sense of these returns for the sample as a whole, let's delete the outlier values realized on April 20-22, 2020 and take some other views. 


The first chart below shows a time series of the excess returns. Even without the April outliers, the interval of March to May 2020 was especially volatile. The second chart below demonstrates this elevated volatility, by showing the 60 day rolling standard deviation of the excess return. 


The third chart below shows a histogram of the excess returns, excluding the April 20-22 outliers. Of course the outlier selection is somewhat arbitrary and impacts the shape of the histogram. That said, in terms of central tendency, the mean daily excess return is 0.23 percent, while the median is 0.09 percent. The histogram demonstrates that the number of daily excess returns above 15% is greater than the number of daily excess returns below -15%.

```python
series.loc['2020-04-20':'2020-04-22'] = np.nan
series = series.dropna()

print('Difference in Daily Percent Change in Index\nOil less 3M UST\nExcl. April 20 - April 22, 2020')
print('Median: ', np.round(series.median(), 3), ' ; Mean: ', np.round(series.mean(), 3))
      

lineChart(series, 'Difference in Daily Percent Change in Index\nOil less 3M UST\nExcl. April 20 - April 22, 2020')


std_dev = series.rolling(60).std().dropna()
title = 'Excess Return\n60 Day Trailing Volatility\nExcl. April 20 - April 22, 2020'
lineChart(std_dev, title, wide=True)

fig, ax = plt.subplots()
h = ax.hist(series, bins=50)
ax.set_title('Difference in Daily Percent Change in Index\nOil less 3M UST\nExcl. April 20 - April 22, 2020')
plt.show()
plt.close()
```

## (B) Log Cumulative Excess Return


Log cumulative excess returns are found by separately calculating log cumulative returns for oil and UST, then taking the difference. 

Log cumulative returns are calculated by first scaling the indexes such that December 31, 2019 = 1. Then, the log cumulative returns are equal to the log of the re-scaled index. This is consistent with the interpretations of the two indexes discussed above, i.e. that the percent change in each index over time reflects the return from holding a given position. 

Finally, log cumulative excess returns are found by taking the log cumulative returns for oil spot prices, and subtracting out the log cumulative returns for 3M UST. 


From December 31, 2019 to December 31, 2020, the log cumulative excess return over the period reaches -23.8. This negative figure is largely driven by the decline in spot oil prices over the period, to USD48.35/bl as of 12/31/2020 from USD61.14/bl as of 12/31/2019, a decline of 20.9%. 

```python

## log returns
df = pd.concat([oil['OIL_index'], ust['DTB3_index']], 1)
df = df.loc['2019-12-31':'2020-12-31']
df = np.log(df / 100) * 100

twoLineChart(df.loc['2019-12-31':'2020-12-31','DTB3_index'], '3M UST', 
             df.loc['2019-12-31':'2020-12-31','OIL_index'], 'Oil',  
             'Cumulative Change in Log of Index\nOil & 3M UST')

## (B) log cumulative excess return
series = df.loc['2019-12-31':'2020-12-31','OIL_index'] - df.loc['2019-12-31':'2020-12-31','DTB3_index']
lineChart(series, 'Difference in Cumulative Change in Log of Index\nOil less 3M UST')
```
